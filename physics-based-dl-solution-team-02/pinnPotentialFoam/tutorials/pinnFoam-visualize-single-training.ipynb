```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import rcParams
import os
rcParams["figure.dpi"] = 200

data_frame = pd.read_csv("cylinder/pinnPotentialFoam-00000000.csv")
plt.plot(data_frame["EPOCH"], data_frame["DATA_MSE"], label="DATA_MSE")
plt.plot(data_frame["EPOCH"], data_frame["RESD_MSE"], label="RESD_MSE")
plt.plot(data_frame["EPOCH"], data_frame["TRAINING_MSE"], label="TRAINING_MSE")
plt.loglog()
plt.legend()
```
